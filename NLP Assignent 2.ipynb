```python
import re
```

```python
def getDate(string):
    dts=[]
    a1=re.findall(r'[0-2][0-9]\sJanuary\s\d{4}',string)
    if a1:
        dts.append(a1[0])
        
    a2=re.findall(r'3[0-1]\sJanuary\s\d{4}',string)
    if a2:
        dts.append(a2[0])
    
    a3=re.findall(r'[0-2]?[0-9]\sFebruary\s\d{4}',string)
    if a3:
        dts.append(a3[0])
    
    #a4=re.findall(r'3[0-1]\sFebruary\s\d{4}',string)
    #if a4:
        #dts.append(a4[0])
    
    a5=re.findall(r'[0-2][0-9]\sMarch\s\d{4}',string)
    if a5:
        dts.append(a5[0])
        
    a6=re.findall(r'3[0-1]\sMarch\s\d{4}',string)
    if a6:
        dts.append(a6[0])
        
    a7=re.findall(r'[0-2][0-9]\sApril\s\d{4}',string)
    if a7:
        dts.append(a7[0])
     
    a8=re.findall(r'30\sApril\s\d{4}',string)
    if a8:
        dts.append(a8[0])
     
    a9=re.findall(r'[0-2][0-9]\sMay\s\d{4}',string)
    if a9:
        dts.append(a9[0])
    
    a10=re.findall(r'3[0-1]\sMay\s\d{4}',string)
    if a10:
        dts.append(a10[0])    
    
    a11=re.findall(r'[0-2][0-9]\sJune\s\d{4}',string)
    if a11:
        dts.append(a11[0])
        
    a12=re.findall(r'30\sJune\s\d{4}',string)
    if a12:
        dts.append(a12[0]) 

    a13=re.findall(r'[0-2][0-9]\sJuly\s\d{4}',string)
    if a13:
        dts.append(a13[0])
    
    a14=re.findall(r'3[0-1]\sJuly\s\d{4}',string)
    if a14:
        dts.append(a14[0])
        
    a15=re.findall(r'[0-2][0-9]\sAugust\s\d{4}',string)
    if a15:
        dts.append(a15[0])
     
    a16=re.findall(r'3[0-1]\sAugust\s\d{4}',string)
    if a16:
        dts.append(a16[0])
        
    a17=re.findall(r'[0-2][0-9]\sSeptember\s\d{4}',string)
    if a17:
        dts.append(a17[0])
        
    a18=re.findall(r'30\sSeptember\s\d{4}',string)
    if a18:
        dts.append(a18[0])
     
    a19=re.findall(r'[0-2][0-9]\sOctober\s\d{4}',string)
    if a19:
        dts.append(a19[0])
        
    a20=re.findall(r'3[0-1]\sOctober\s\d{4}',string)
    if a20:
        dts.append(a20[0])
    
    a21=re.findall(r'[0-2][0-9]\sNovember\s\d{4}',string)
    if a21:
        dts.append(a21[0])
    
    a22=re.findall(r'30\sNovember\s\d{4}',string)
    if a22:
        dts.append(a22[0])
        
    a23=re.findall(r'[0-2][0-9]\sDecember\s\d{4}',string)
    if a23:
        dts.append(a23[0])
        
    a24=re.findall(r'3[0-1]\sDecember\s\d{4}',string)
    if a24:
        dts.append(a24[0])
        
        
    a25=re.findall(r'[0-2][0-9]/0?[1-9]/\d{4}',string)
    if a25:
        dts.append(a25[0])
        
    
    a26=re.findall(r'3[0-1]/0?[1-9]/\d{4}',string)
    if a26:
        dts.append(a26[0])
        
    a27=re.findall(r'[0-2][0-9]/1[0-2]/\d{4}',string)
    if a27:
        dts.append(a27[0])
        
    a28=re.findall(r'3[0-1]/1[0-2]/\d{4}',string)
    if a28:
        dts.append(a28[0])
        
    a29=re.findall(r'[0-2][0-9]-0?[1-9]-\d{4}',string)
    if a29:
        dts.append(a29[0])
        
    a30=re.findall(r'3[0-1]-0?[1-9]-\d{4}',string)
    if a30:
        dts.append(a30[0])
        
    a31=re.findall(r'[0-2][0-9]-1[0-2]-\d{4}',string)
    if a31:
        dts.append(a31[0])  
        
    a32=re.findall(r'3[0-1]-1[0-2]-\d{4}',string)
    if a32:
        dts.append(a32[0])
     
    a33=re.findall(r'[0-2][0-9]\.0?[1-9]\.\d{4}',string)
    if a33:
        dts.append(a33[0])
    
    a34=re.findall(r'3[0-1]\.0?[1-9]\.\d{4}',string)
    if a34:
        dts.append(a34[0])
        
    a35=re.findall(r'[0-2][0-9]\.1[0-2]\.\d{4}',string)
    if a35:
        dts.append(a35[0])

    a36=re.findall(r'3[0-1]\.1[0-2]\.\d{4}',string)
    if a36:
        dts.append(a36[0])
        
    if not dts:
        print("No date Found")
        #break
    
    return dts
```

```python
#Correct Dates passed to function
print(getDate("15 November 1989"))
print(getDate("26/11/2017"))
print(getDate("15.09.1990"))
print(getDate("27-09-1991"))
print(getDate("12/1/2019"))
print(getDate("01/01/2012"))
print(getDate("01.07.1990"))
```

```python
#Incorrect Dates passed to function
print(getDate("32 November 1989"))
print(getDate("32 Sachin 1989"))
print(getDate("1-15-1989"))
print(getDate("1.1.12"))
print(getDate("04@04@2017"))
```
